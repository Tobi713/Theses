```python
###Distributions for boundary locations
alpha = pm.Normal("alpha", -2000, 1./np.square(7.), value= 5)
beta = pm.Normal("beta", -2050, 1./np.square(25), value=-500)
gamma = pm.Normal("gamma", -2200, 1./np.square(45), value = 0)


###Model
model = pm.Model([alpha, beta, gamma])
```

```python
###PyMC simulation
M = pm.MCMC(model)
M.sample(iter=5000)
```

```python
import scipy.optimize as sop

#Loss function for reservoir thickness.
def res_thickness_loss(estimate_th, true_th, risk_th, flex=1500):
        loss_th = np.zeros_like(true_th)
        buffer = [abs(true_th - estimate_th) <= (1/risk_th)*flex] 
        #Flexibility buffer allows adaption slightly higher or lower values. Flexibility depends on risk.
        underestim = estimate_th < true_th
        loss_th[underestim] = np.abs(estimate_th - true_th[underestim]) + 2*risk_th
        #If thickness is underestimated, costs arise due to adaptions to capture higher OOIP.
        loss_th[~underestim] = np.abs(estimate_th - true_th[~underestim])*(risk_th/10) + risk_th
        #If thickness is overestimated, we invest more than needed and capture less than expected.
        loss_th[buffer] = risk_th
        return loss_th/10000
    
#def res_depth_loss(estimate_dp, true_dp, risk_dp, alpha=100.):
#        if estimate_dp < 
#        loss = (estimate_dp ** 2)/alpha
#        return loss_dp
    
def seal_thickness_loss(estimate_st, true_st, risk_st):
        loss_st = np.zeros_like(true_st)
        seal_break = risk_st
        brk = estimate_st < seal_break
        safe = [estimate_st > (risk_st*2)]
        loss_st[safe] = 1
        loss_st[~safe] = 1 - (estimate_st * (1/true_st[~safe]))
        return loss_st/10
```

```python
figsize(15, 6)

estimates_th = np.arange(0,400,5)
risks_th = np.arange(20,60,10)
  
expected_loss_th = lambda estimate_th, risk_th: res_thickness_loss(estimate_th, reservoir_thickness, risk_th).mean()

plt.hist((reservoir_thickness), histtype='stepfilled', bins=30, alpha=1,
    label="Reservoir", normed=True, color = "g")

for r in risks_th:
    #cut_off = 35
    loss = [expected_loss_th(e, r) for e in estimates_th]
    _min_results = sop.fmin(expected_loss_th, 40, args=(r,),disp = False)
    plt.plot(estimates_th, loss, label = "Risk r "+str(r))
    plt.scatter(_min_results, 0, s = 20, color = 'black', label = "Bayes action r "+str(r))
    plt.vlines(_min_results, 0, 0.02, color = 'black', linestyles="--")
    print("Bayes action (minimum) at risk r %d: %.2f" % (r, _min_results))

plt.title(r"""Expected loss relative to reservoir thickness""")    
plt.legend(loc="upper left", scatterpoints=1, title="Legend")
plt.xlabel("Reservoir thickness estimate")
plt.ylabel("Expected loss")
#plt.xlim(-20, 400)
#plt.ylim(0,0.02)
plt.grid()
```

```python
#MCMC with 2 inputs: reservoir thickness and depth
#true_quality = pm.Normal("true_quality", 2000, 1.0 / 1000 ** 2) #true_quality prior?

th_v = 20 #value per meter reservoir thickness
dp_c = 1  #cost per meter depth

res_top = pm.Normal("res_top", -2050, 1./np.square(25), value=-500)
res_bottom = pm.Normal("res_bottom", -2200, 1./np.square(45), value = 0)

@pm.deterministic
def res_thickness(res_top=res_top, res_bottom=res_bottom):
    return np.abs(res_top - res_bottom)

@pm.deterministic
def true_quality(res_thickness=res_thickness, res_depth=res_top, th_v=th_v, dp_c=dp_c):
    return th_v * res_thickness + (dp_c * res_depth) 

model = pm.Model([true_quality, res_thickness, res_top, res_bottom])

M = pm.MCMC(model)
M.sample(iter=5000)
```

```python
seal_top = pm.Normal("seal_top", -2000, 1./np.square(7.))

seal_to = 20 #Minimum treshold for seal to be safe

@pm.deterministic
def value_if_seal(res_thickness=res_thickness, res_depth=res_top, \
                  th_v=th_v, dp_c=dp_c):
    return (th_v * res_thickness + (dp_c * res_depth))

@pm.deterministic
def true_quality(seal_top=seal_top, res_top=res_top,
                 dp_c=dp_c, res_depth=res_top, 
                 value_if_seal = value_if_seal):
    #print(seal_top, res_top)
    seal_th = seal_top - res_top
    if seal_th >= seal_to:
        #print (seal_th)
        x = value_if_seal
    else:
        x = (dp_c * res_depth)
  #  print (x)
    return x


model = pm.Model([true_quality, res_thickness, value_if_seal, res_top, res_bottom, seal_top])

M = pm.MCMC(model)
M.sample(iter=5000)
```

## Designing a loss function

Based on the distribution for the reservoir quality derived using three parameters from above, it is possible to design a loss function to express the expected loss of possible decisions (based on predictions/estimates) and represent the variations in expected loss for decision-makers with different risk-affinities.

A company interested in developing such a reservoir wants to predict the quality of this reservoir, i.e. the value present. A prediction should be as precise as possible, so as not to allocate too many or too few resources to the project.

Proposed loss function, disregarding risk:

$$L( \theta, \hat{\theta} ) = \begin{cases} \theta & \theta \gt 0, \hat{\theta} \le 0 \\\ \hat{\theta} & \theta \le 0, \hat{\theta} \gt 0 \\\ 0 & \theta \lt 0, \hat{\theta} \lt 0 \\\ |( \theta -  \hat{\theta} )| \end{cases}$$

Proposed loss function, including risk (defined by risk factor r):

$$L( \theta, \hat{\theta} ) = \begin{cases} r \theta & \theta \gt 0, \hat{\theta} \le 0 \\\ r \hat{\theta} & \theta \le 0, \hat{\theta} \gt 0 \\\ 0 & \theta \lt 0, \hat{\theta} \lt 0 \\\ |( \theta -  \hat{\theta} )| \end{cases}$$
